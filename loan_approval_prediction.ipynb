```python
# for dataset d1
```

```python
x1=d1.drop(columns="loan_status")
y1=d1.iloc[:,-1]
 
```

```python
x1
```

```python
from sklearn.preprocessing import StandardScaler
s=StandardScaler()
x1=s.fit_transform(x1)
```

```python
from sklearn.model_selection import train_test_split
x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,test_size=0.2,random_state=42)
```

```python
print("x1_train.shape",x1_train.shape)
print("x1_test.shape",x1_test.shape)
print("y1_train.shape",y1_train.shape)
print("y1_test.shape",y1_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
param_grid = [
    {'penalty':['l1','l2','elasticnet','none'],
    'C' : np.logspace(-4,4,20,base=10),
    'solver': ['lbfgs','newton-cg','liblinear','sag','saga'],
    'max_iter'  : [100,1000,2500,5000]
}
]
```

```python
from sklearn.model_selection import GridSearchCV
clf=GridSearchCV(estimator=model,param_grid=param_grid,cv=5,n_jobs=-1)
```

```python
best_clf=clf.fit(x1_train,y1_train)
best_clf.best_params_
```

```python
best_clf.best_params_
```

```python
from sklearn.metrics import accuracy_score,classification_report
y1_pred=best_clf.predict(x1_test)
print(f'Training Accuracy  : {best_clf.score(x1_train,y1_train):.3f}')
print("Test Accuracy :",accuracy_score(y1_test,y1_pred))
print("-------------------------------------------------------------------------------")
print(classification_report(y1_test,y1_pred))
```

```python
x2=d2.drop(columns="loan_amount")
y2=d2.iloc[:,4]
```

```python
x2
```

```python
y2
```

```python
from sklearn.preprocessing import StandardScaler
s2=StandardScaler()
x2=s2.fit_transform(x2)
```

```python
x2
```

```python
y2=y2/100000000
```

```python
y2
```

```python
x2_train,x2_test,y2_train,y2_test=train_test_split(x2,y2,test_size=0.2,random_state=44)
print("x2_train.shape",x2_train.shape)
print("x2_test.shape",x2_test.shape)
print("y2_train.shape",y2_train.shape)
print("y2_test.shape",y2_test.shape)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
param_grid = {
    'poly_features__degree': [1,2,3],  # Polynomial degrees to test
    'regressor': [LinearRegression()]  # Only Polynomial Regression
}
```

```python
pipeline = Pipeline([
    ('poly_features', PolynomialFeatures()),  # Generates polynomial terms
    ('regressor', LinearRegression())  # Linear Regression model
])
```

```python
clf2 = GridSearchCV(pipeline, param_grid, cv=5, scoring='r2')
 
```

```python
clf2.fit(x2_train,y2_train)
```

```python
print("\nBest parameters:", clf2.best_params_)
print("Best R² score on training set:", clf2.best_score_)

# Evaluate on test set
best_model = clf2.best_estimator_
test_score = best_model.score(x2_test, y2_test)
print("Test R² score:", test_score)
```

```python
y2_pred=clf2.predict(x2_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
# Calculate Metrics
mae = mean_absolute_error(y2_test, y2_pred)
mse = mean_squared_error(y2_test, y2_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y2_test, y2_pred)

# Print Results
print(f"Mean Absolute Error (MAE): {mae:.3f}")
print(f"Mean Squared Error (MSE): {mse:.3f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.3f}")
print(f"R² Score: {r2:.3f}")
```
